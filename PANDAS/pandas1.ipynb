```python
df1 = pd.DataFrame({'roll':[1,2,3,4], 'maths':[10,12,13,14], 'physics':[13,12,12,13]})
df2 = pd.DataFrame({'roll':[1,2,5,6], 'maths':[10,11,14,10], 'chemistry':[13,10,11,13]})
```

```python
pd.merge(df1,df2) #merging done on common elements from common columns
```

```python
pd.merge(df1,df2,how='left')
```

```python
pd.merge(df1,df2,how='right')
```

```python
pd.merge(df1,df2,how='outer')
```

```python
df1 = pd.DataFrame({'roll':[1,2,3,4], 'maths':[10,12,13,14], 'physics':[13,12,12,13]})
df2 = pd.DataFrame({'roll':[1,2,5,6], 'maths':[10,11,14,10], 'chemistry':[13,10,11,13]})

pd.concat([df1,df2], ignore_index=True, sort=True)
```
