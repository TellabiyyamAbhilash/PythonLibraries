```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
sns.set_style("white")
import random

sns.__version__
```

LINE PLOT

```python
roll = [i for i in range(1,21)]
marks = [random.randint(30,90) for i in range(1,21)]

sample_df = pd.DataFrame({'roll':roll,'marks':marks})
```

```python
sns.lineplot(x='roll', y='marks', data = sample_df )
plt.title('Student Marks')
plt.figure(figsize=(20,20))
```

```python
planets_df = sns.load_dataset('planets')
planets_df.head()
```

```python
plt.figure(figsize=(12,6))
bins = [2004,2005,2006,2007,2008,2009,2010,2011]
sns.lineplot(x = 'year', y = 'distance', data = planets_df, palette = ['red','blue'], hue='method', hue_order=['Astrometry','Radical Velocity',  'Transit'])
plt.xticks(bins)
```
