```python
import seaborn as sns
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
df = sns.load_dataset('penguins')
df.head()
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='species')
```

```python
sns.pairplot(df, markers = ["o","s", "D"], hue = "species", palette = "rainbow")
```

```python
sns.pairplot(df, markers = ["o","s", "D"], hue = "species", palette = "rainbow", corner = True)
```
